```python
def save_counter_bigram(m = 0):
    n = 2
    counter = []
    start_l = ['<s>']*(n-1)
    start = '<s>\u0020'*(n-1)
    i = 0
    
    
    #only saving counters from 300,000 lines at a time
    for line in open('data/compiled.txt', encoding='utf-8'):
        if(300000*m<=i<=300000*(m+1)):
            line = re.sub(r'\s+[\u0964]\s+', r'\u0020\u0964\u0020'+start, line)
            #text = re.findall(r'[\u0900-\u097F]+|\u003c\u002f\u0073\u003e|\u003c\u0073\u003e',text.lower().strip())
            line = re.sub(r'p', r's', line)

            c = ([tuple(x.split()) for x in re.findall(r'\s+[\u0900-\u097F]+\s+[\u0900-\u097F]+|\u003c\u0073\u003e\s+[\u0900-\u097F]+|[\u0900-\u097F]+\s+\u003c\u0073\u003e', line, overlapped=True)])      
            counter+=c
            i+=1
        elif (i<=300000*m):
            i+=1
#         if(i%100==0):
#             print(i)
        if (i > 300000*(m+1)):
            break
    return counter

```
