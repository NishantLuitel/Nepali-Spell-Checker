```python





# output = reader.readtext(loc)
prefix_name = "data/ilovepdf_pages-to-jpg/Sabdkosh-2075-10th_page-"
loc_num = 0
postfix_name = ".jpg"


txt = ''
for i in range(1380):
    loc_num = i+1
    if loc_num < 10:
        loc_str = '000'+str(loc_num)
    elif loc_num >= 10 and loc_num < 100:
        loc_str = '00'+str(loc_num)
    elif loc_num >= 100 and loc_num < 1000:
        loc_str = '0'+str(loc_num)
    else:
        loc_str = str(loc_num)
    filename = prefix_name + loc_str + postfix_name
    
    output = reader.readtext(filename)
    
    tr = [output[i][1] for i in range(len(output))]

    txt += ' '.join(tr)
    
    if i%10 ==0:
        print(i+1)




# Open the file in write mode ('w')
with open(file_path, 'w',encoding = 'utf-8') as file:
    # Write the string to the file
    file.write(txt)
```

```python
import regex as re
```

```python
with open(file_path, 'r',encoding = 'utf-8') as file:
    # Write the string to the file
    content = file.read()
```

```python

# From the dictionary match all the nepali vocab that ends with '-' so that 
# the word is likely the one whose meaning is mentioed

def words(text): 
    text = re.sub(r'\u0945', r'\u0901', text)
    text = re.sub(r' -', r'-', text)
    text = re.sub(r'\s{2,}', r'- ', text)
    text = re.sub(r'--', r'-', text)
    text = re.sub(r'-\s{2,}', r'- ', text)
    return re.findall(r'[\u0900-\u0950]+- ', text.lower())
```

```python
# Extract the token$ and remove duplicate$

lexicons = words(content)
lexicons = list(set(lexicons))
len(lexicons)
```

```python
sorted(lexicons)[5200:]
```

```python
# Format more

lexicons2 = [word[:-2] for word in lexicons]
lexicons2 = sorted(lexicons2)

# Word that contain bindu in our extraction
bindu_lexicons = [word for word in lexicons2 if '' in word or '' in word]
len(bindu_lexicons)
```
