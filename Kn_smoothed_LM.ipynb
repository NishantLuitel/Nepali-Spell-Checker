```python
import math
def perplexity(mdl,sent):
    sentence_len = len(sent)
    sent = ['<s>'] * (mdl.highest_order - 1) + sent      
    logprob = mdl.score_sent(tuple(sent))
    print(logprob)
    per = math.exp(-logprob/sentence_len)
    return per



```

```python
def get_context(sentence,order):
    return sentence[(len(sentence) - order + 1):]
```

```python
import math
import random
#Generating word without random element makes the model repetitive 
def generate_next_word(mdl,sent):

        probs=mdl.lm[0]
        context = tuple(get_context(sent,mdl.highest_order))
        print('Context:',context)
        pos_ngrams = [
            (ngram, logprob) for ngram, logprob in probs.items()
            if list(ngram)[:-1] == list(context)]
        
        if len(pos_ngrams) == 0:
            print("Backed Off")
            pos_ngrams = [(ngram, logprob) for ngram, logprob in mdl.lm[1].items()
            if list(ngram)[:-1] == list(context[1:])]
            probs = mdl.lm[1]
            
        # Normalize to get conditional probability.
        # Subtract max logprob from all logprobs to avoid underflow.
        _, max_logprob = max(pos_ngrams, key=lambda x: x[1])
        return _[-1]
```

```python
#When random element is present the model makes more diversed prediction
def generate_next_word(mdl,sent):

        probs=mdl.lm[0]
        context = tuple(get_context(sent,mdl.highest_order))
        print('Context:',context)
        pos_ngrams = [
            (ngram, logprob) for ngram, logprob in probs.items()
            if list(ngram)[:-1] == list(context)]
        
        if len(pos_ngrams) == 0:
            print("Backed Off")
            pos_ngrams = [(ngram, logprob) for ngram, logprob in mdl.lm[1].items()
            if list(ngram)[:-1] == list(context[1:])]
            probs = mdl.lm[1]
            
        # Normalize to get conditional probability.
        # Subtract max logprob from all logprobs to avoid underflow.
        _, max_logprob = max(pos_ngrams, key=lambda x: x[1])
        pos_ngrams = list(
            (ngram, math.exp(prob - max_logprob)) for ngram, prob in pos_ngrams)
        total_prob = sum(prob for ngram, prob in pos_ngrams)
        pos_ngrams = list(
            (ngram, prob/total_prob) for ngram, prob in pos_ngrams)
        rand = random.random()
        for ngram, prob in pos_ngrams:
            rand -= prob
            if rand < 0:
                return ngram[-1]
        return ngram[-1]





x = 'बृहत रूपमा प्रयोग गरिएको गणित'
for i in range(5):
    m = generate_next_word(kn_lm2,x.split())
    x+=' '+m
print(x)
```
